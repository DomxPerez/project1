```python
import csv
import random
from datetime import datetime, timedelta

# Define the Product class
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

# Define the Order class
class Order:
    def __init__(self, products):
        self.products = products

# Define the Customer class
class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name

    def create_order(self, products):
        return Order(products)

# Define the Store class
class Store:
    def __init__(self, store_id):
        self.store_id = store_id
        self.customers = []

    def add_customer(self, customer):
        self.customers.append(customer)

# Define the Corporation class
class Corporation:
    def __init__(self, name):
        self.name = name
        self.stores = []

    def add_store(self, store):
        self.stores.append(store)

# Generate products
products = [
    Product("Laptop", 1200),
    Product("Smartphone", 800),
    Product("Tablet", 400),
    Product("Headphones", 150),
    Product("Smartwatch", 200),
    Product("Monitor", 300),
    Product("Keyboard", 100),
    Product("Mouse", 50),
    Product("Printer", 250),
    Product("Camera", 500)
]

# Generate customers
customer_names = ["Alice", "Bob", "Charlie", "David", "Eve"]
customers = [Customer(i, name) for i, name in enumerate(customer_names)]

# Generate stores and add customers to stores
stores = [Store(i) for i in range(3)]
for store in stores:
    for customer in customers:
        store.add_customer(customer)

# Create a corporation and add stores to it
corporation = Corporation("Dom Enterprise")
for store in stores:
    corporation.add_store(store)

# Generate sales data and write to CSV file
with open('sales_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Time", "StoreID", "CustomerID", "Customer Name", "OrderID", "Product Name", "Price"])

    order_id = 1
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 12, 31)

    for _ in range(1000):  # Generate 1000 orders
        store = random.choice(corporation.stores)
        customer = random.choice(store.customers)

        order_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
        order_time = order_date + timedelta(hours=random.randint(0, 23), minutes=random.randint(0, 59))
        order_products = random.sample(products, random.randint(1, len(products)))
        order = customer.create_order(order_products)

        for product in order.products:
            writer.writerow([order_date.date(), order_time.time(), store.store_id, customer.customer_id, customer.name, order_id, product.name, product.price])

        order_id += 1

print("The sales data has been written to 'sales_data.csv'.")
```
